1. В поезде 9 вагонов. Сколькими способами можно посадить в него 4 человека так, чтобы все они оказались в различных вагонах?

**Решение**

$ A^n_k= \frac {n!} {(n-k)!} = \frac {9!}{5!} = 9\cdot 8\cdot 7 \cdot 6 = 3024$

```python
import math

n = 9  # Число вагонов
k = 4  # Число человек

# Число размещений без повторений
count = math.perm(n, k)
print(count) 
```

2. В России автомобильный номерной знак состоит из трех цифр и трех букв. При этом номер, содержащий три нуля, недопустим, и буквы выбираются из набора А, В, Е, К, М, Н, О, Р, С, Т, У, Х. Сколько есть всего различных автомобильных номеров?

**Решение**

Число размещений с повторениями: $n^k$

Вариантов для цифр: $10^3 - 1 = 999$ <br>
Вариантов для букв: $12^3 = 1728$ <br>
Всего вариантов: $999\cdot1728 = 1726272$

```python
d_count = 10
d_posit = 3
l_count = 12
l_posit = 3

d_options = d_count**d_posit - 1 # Вариантов для цифр
l_options = l_count**l_posit # Вариантов для букв

t_options = d_options*l_options
print (t_options)
```
