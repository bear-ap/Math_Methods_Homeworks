## 2.1. Случайная величина и распределение

Случайная величина $\xi $ имеет абсолютно непрерывное распределение на $\mathbb{R}$, его плотность задана формулой $p(x)=Cx^{-3/2}\mathbb{I}_{[1,\infty )}$. <br>

1. Найти постоянную $C$
2. Найти $\mathbb{P} \left( \frac{1}{10} < \frac{1}{\xi} < \frac{1}{5} \right)$, ответ округлить до четвертого знака после запятой.


```python
# Импорт библиотеки sympy
import sympy as sp

# Определяем переменные
x = sp.Symbol('x', real=True)
c = sp.Symbol('c', real=True)

# 1. Находим константу C
# Плотность задана как C * x**(-3/2) на интервале [1, +inf)
density_expr = c * x**(-sp.Rational(3, 2))

# Условие нормировки: интеграл плотности по всей прямой равен 1
integral_eq = sp.integrate(density_expr, (x, 1, sp.oo))
c_value = sp.solve(sp.Eq(integral_eq, 1), c)[0]

print(f"1. {c_value}")

# 2. Находим P
lower_limit = 5
upper_limit = 10

# Подставляем найденное значение C в выражение плотности
p_x = density_expr.subs(c, c_value)

# Вычисляем вероятность через интеграл
probability = sp.integrate(p_x, (x, lower_limit, upper_limit))
prob_numeric = float(probability.evalf())

print(f"2. {round(prob_numeric,4)}")
```

## 2.3. Закон больших чисел

### Задача 1 
Пусть случайная величина $\eta _{n}$ равна сумме очков, появившихся при $n$ независимых подбрасываниях честной игральной кости. <br>
Используя неравенство Чебышева, оценить вероятность$P\left(\left|\frac{\eta _{n}}{n}-3.5\right|>\varepsilon \right)\le \dots $Ответ приведите для $n = 10, \varepsilon = 1$

```python
def chebyshev_estimate(n, eps):
    # Дисперсия одной кости: (35 / 12)
    var_single = 35 / 12
    
    # Дисперсия среднего арифметического n костей
    var_mean = var_single / n
    
    # Оценка по неравенству Чебышёва: P > eps <= Var / eps^2
    probability_limit = var_mean / (eps ** 2)
    
    return probability_limit

n = 10
eps = 1
result = chebyshev_estimate(n, eps)

print(f"Для n = {n} и epsilon = {eps}:")
print(f"Оценка вероятности P <= {result:.4f} (или {7}/24)")
```

### Задача 2
Удовлетворяет ли закону больших чисел последовательность независимых случайных величин $(\xi_k)_{k=1}^\infty$, если$P(\xi _{k}=\sqrt{k})=P(\xi _{k}=-\sqrt{k})=\frac{1}{2\sqrt{k}},\quad P(\xi _{k}=0)=1-\frac{1}{\sqrt{k}}?$ <br>
Указание: возможно, вам понадобится закон больших чисел в форме Маркова. Погуглите!

### Решение

**Шаг 1. Нахождение математического ожидания $M(\xi_k)$** <br>
$$\xi _{k}$:$M(\xi _{k})=\sqrt{k}\cdot \frac{1}{2\sqrt{k}}+(-\sqrt{k})\cdot \frac{1}{2\sqrt{k}}+0\cdot \left(1-\frac{1}{\sqrt{k}}\right)=\frac{1}{2}-\frac{1}{2}+0=0$$
Так как $M(\xi_k) = 0$ для всех $k$, то:
$$\frac{1}{n}\sum _{k=1}^{n}M(\xi _{k})=0$$

**Шаг 2. Нахождение дисперсии $D(\xi_k)$** <br> 
Т.к. математическое ожидание равно нулю, $D(\xi_k) = M(\xi_k^2) - (M(\xi_k))^2 = M(\xi_k^2)$:
$$D(\xi _{k})=(\sqrt{k})^{2}\cdot \frac{1}{2\sqrt{k}}+(-\sqrt{k})^{2}\cdot \frac{1}{2\sqrt{k}}+0^{2}\cdot \left(1-\frac{1}{\sqrt{k}}\right)$$
$$D(\xi _{k})=k\cdot \frac{1}{2\sqrt{k}}+k\cdot \frac{1}{2\sqrt{k}}=\frac{k}{\sqrt{k}}=\sqrt{k}$$

**Шаг 3. Проверка условия Маркова (Чебышёва)** <br>
Закон больших чисел в форме Маркова утверждает, что если для последовательности случайных величин выполняется условие:
$$\lim _{n\rightarrow \infty }\frac{1}{n^{2}}D\left(\sum _{k=1}^{n}\xi _{k}\right)=0,$$ 
то к этой последовательности применим закон больших чисел. 

Так как случайные величины по условию независимы, дисперсия суммы равна сумме дисперсий (это частный случай теоремы Маркова, называемый теоремой Чебышёва). Проверим предел:
$$\lim _{n\rightarrow \infty }\frac{1}{n^{2}}\sum _{k=1}^{n}D(\xi _{k})=\lim _{n\rightarrow \infty }\frac{1}{n^{2}}\sum _{k=1}^{n}\sqrt{k}$$

Поскольку каждый член $\sqrt{k} \leq \sqrt{n}$ при $k \leq n$, то вся сумма из $\sum _{k=1}^{n}\sqrt{k}\le n^{3/2}$

Подставим эту оценку в предел:
$$0\le \lim _{n\rightarrow \infty }\frac{1}{n^{2}}\sum _{k=1}^{n}\sqrt{k}\le \lim _{n\rightarrow \infty }\frac{n^{3/2}}{n^{2}}=\lim _{n\rightarrow \infty }\frac{1}{\sqrt{n}}=0$$

По теореме о зажатой функции предел строго равен 0.

**Вывод** <br>
Т.к. условие Маркова (Чебышёва) выполнено, среднее арифметическое данных случайных величин сходится по вероятности к нулю при $n \to \infty$: $$\frac{\xi _{1}+\xi _{2}+\dots +\xi _{n}}{n}\xrightarrow{P}0$$
Следовательно, последовательность удовлетворяет закону больших чисел. 

## 2.4. Случайные векторы. Ковариация и корреляция. Задача линейной регрессии.
1. Честная игральная кость подбрасывается  $n $ раз. Пусть  $\xi  $ есть число выпаданий единицы, а  $\eta  $ — число выпаданий шестерки. Найти коэффициент корреляции  $\rho(\xi, \eta) $.


```python
def calculate_correlation():
    n = sp.symbols('n', positive=True, integer=True)
    p = sp.Rational(1, 6)
    
    # Дисперсия биномиального распределения
    var_xi = n * p * (1 - p)
    var_eta = n * p * (1 - p)
    
    # Ковариация для мультиномиального распределения: -n * p1 * p2
    cov_xi_eta = -n * p * p
    
    # Коэффициент корреляции
    rho = cov_xi_eta / sp.sqrt(var_xi * var_eta)
    
    return sp.simplify(rho)

result = calculate_correlation()
print(f"Коэффициент корреляции rho(xi, eta) = {result}")
# Вывод: -1/5
```

2. Случайные величины  $\xi_1, \xi_2, \xi_3, \xi_4 $ независимы в совокупности. Верно ли, что случайные векторы  $\eta = (\xi_1, \xi_2) $ и  $\zeta = (\xi_3, \xi_4) $ независимы? Верно ли обратное?

### Решение
1. Из независимости величин следует независимость векторов

Если случайные величины $\xi_1, \xi_2, \xi_3, \xi_4$ независимы в совокупности, то любые непересекающиеся группы этих величин также независимы.

Случайные векторы $\eta = (\xi_1, \xi_2)$ и $\zeta = (\xi_3, \xi_4)$ сформированы из разных (непересекающихся) наборов независимых величин. Согласно свойствам независимости, функции от независимых групп случайных величин сами являются независимыми.

2. Обратное верно.
Если векторы $\eta = (\xi_1, \xi_2)$ и $\zeta = (\xi_3, \xi_4)$ независимы, и при этом внутри самих векторов величины $(\xi_1, \xi_2)$ и $(\xi_3, \xi_4)$ также независимы между собой, то вся совокупность $\xi_1, \xi_2, \xi_3, \xi_4$ будет независима в совокупности.
