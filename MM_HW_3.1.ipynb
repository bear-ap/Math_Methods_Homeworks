## 1.1. Комбинаторика

### Задача 1

В поезде 9 вагонов. Сколькими способами можно посадить в него 4 человека так, чтобы все они оказались в различных вагонах?

### Решение

$ A^n_k= \frac {n!} {(n-k)!} = \frac {9!}{5!} = 9\cdot 8\cdot 7 \cdot 6 = 3024$

```python
import math

wag = 9  # Число вагонов
pers = 4  # Число человек

# Число размещений без повторений
count = math.perm(wag, pers)
print(count) 
```

### Задача 2

В России автомобильный номерной знак состоит из трех цифр и трех букв. При этом номер, содержащий три нуля, недопустим, и буквы выбираются из набора А, В, Е, К, М, Н, О, Р, С, Т, У, Х. Сколько есть всего различных автомобильных номеров?

### Решение

Число размещений с повторениями: $n^k$

Вариантов для цифр: $10^3 - 1 = 999$ <br>
Вариантов для букв: $12^3 = 1728$ <br>
Всего вариантов: $999\cdot1728 = 1726272$

```python
d_count = 10 # число цифр
d_posit = 3 # количество цифр в номере
l_count = 12 # число допустимых букв
l_posit = 3 # количество букв в номере

d_options = d_count**d_posit - 1 # Вариантов для цифр
l_options = l_count**l_posit # Вариантов для букв

t_options = d_options*l_options # Всего вариантов
print (t_options)
```

## 1.2. Вероятностное пространство

### Задача 1

Может ли число всех событий какого-либо вероятностного пространства быть равным 65; 66; 64?

### Решение
Множество всех событий конечного вероятностного пространства всегда составляет степень двойки $2^k$.
64 - да, т.к $64 = 2^6$ (пространство из 6 элементарных исходов). 65, 66: Нет, так как эти числа не являются степенями двойки.

```python
def is_power_of_two(n):
    # Число должно быть больше 0, и (n & (n-1)) для степени двойки должно равняться 0
    return n > 0 and (n & (n - 1)) == 0

# Проверка ваших чисел
for num in [64, 65, 66]:
    if is_power_of_two(num):
        print(f"{num} — Да (является степенью двойки)")
    else:
        print(f"{num} — Нет (не является степенью двойки)")
```

 ### Задача 2

Из урны, содержащей m различных шаров, наудачу последовательно вынимают n шаров. Для двух способов выбора, с возвращением и без возвращения, описать структуру пространства элементарных исходов $\Omega$ и подсчитать число элементов $|\Omega|$ в случае <br>
а) упорядоченной выборки <br>
б) неупорядоченной выборки

### Решение

**а) Упорядоченная выборка**
Каждый элементарный исход представляет собой упорядоченную последовательность длины $n$, составленную из номеров извлеченных шаров $(x_1, x_2, \dots, x_n)$, где $x_i \in \{1, 2, \dots, m\}$. <br>
**С возвращением:** <br>
Число элементов: $|\Omega| = m^n$. <br>
**Без возвращения:** <br>
Число элементов (размещения): $|\Omega| = A_m^n = \frac{m!}{(m-n)!}$.

**б) Неупорядоченная выборка**
Порядок появления шаров не учитывается, важен только сам набор из \(n\) выбранных шаров (состав выборки) <br>
**С возвращением:** <br>
Число элементов (сочетания с повторениями): $|\Omega| = \overline{C}_m^n = C_{m+n-1}^n = \frac{(m+n-1)!}{n!(m-1)!}$ <br>
**Без возвращения:** <br>
Число элементов (сочетания): $\|\Omega| = C_m^n = \frac{m!}{n!(m-n)!}$


```python
def omega(m, n):
    print(f"--- Расчет для m = {m}, n = {n} ---")
    
    # а) Упорядоченная выборка
    ord_with = m ** n
    ord_no = math.perm(m, n) if m >= n else 0
    
    # б) Неупорядоченная выборка
    unord_with = math.comb(m + n - 1, n)
    unord_no = math.comb(m, n) if m >= n else 0
    
    print("а) Упорядоченная выборка:")
    print(f"   - С возвращением: {ord_with}")
    print(f"   - Без возвращения: {ord_no}")
    print("б) Неупорядоченная выборка:")
    print(f"   - С возвращением: {unord_with}")
    print(f"   - Без возвращения: {unord_no}\n")

    # Блок ввода данных пользователем
try:
    m = int(input("Введите m (общее количество имеющихся шаров): "))
    n = int(input("Введите n (количество вынимаемых шаров): "))
    
    # Запуск функции с введенными параметрами
    omega(m, n)
except ValueError:
    print("Ошибка: Пожалуйста, вводите только целые числа!")
```

### 1.3. Условная вероятность, независимость и формула Байеса

### Задача 2

Условие: Брошены три игральные кости. Какова вероятность, что на одной из них выпала единица, если известно, что выпали разные числа?

### Решение:

-  Общее число исходов для события, что все три числа разные:
 $ A^n_k= \frac {n!} {(n-k)!} = \frac {6!}{3!} = 6 \cdot 5 \cdot 4 = 120$

- Число благоприятных исходов (все числа разные и среди них есть единица): <br>
Единица может выпасть в 1-й, 2-й или 3-й раз - 3 способа (число сочетаний $C_3^1 = 3$). <br>
Две другие цифры из 5 оставшихся: $A^n_k= \frac {n!} {(n-k)!} = \frac {5!}{3!} = 5 \cdot 4 = 20$ <br>
Общее число таких исходов: $ 3*20 = 60$

- Условная вероятность:
$P= \frac {m}{n} =\frac{60}{120}=0.5$

```python
n = 6
k = 3
g = 1

outcomes_total = math.perm (n,k)

outcomes_given = math.comb (k,g)
outcomes_remained = math.perm (n-g, n-g-k)
outcomes_successful = outcomes_given*outcomes_remained

p_cond = outcomes_successful / outcomes_total

print (p_cond)
```

Условие: Из колоды, состоящей из 36 карт наугад выбирают две карты. С какой вероятностью они окажутся одной масти? Ответ округлить до четвертого знака после запятой.
